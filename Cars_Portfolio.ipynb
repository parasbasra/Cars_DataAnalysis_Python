# Data Analysis with visualization on Cars Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.show()
```

## Loading the dataset

```python
df = pd.read_csv(r'C:\Users\Lenovo\Downloads\cars.csv')
```

## It has 432 rows and 15 columns

```python
df
```

## Showing all the null values and removing it from the dataset

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

## 1- Make Distribution
### With Toyota being the most selling Company and Hummer being the least.

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Make', data=df)
plt.xticks(rotation=90)
plt.title('Distribution of Car Makes')
plt.show()
```

## 2- Type of vehicles
### Sedan being the most common type of vehicle with Hybrid being the least.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Type', data=df)
plt.title('Distribution of Vehicle Types')
plt.show()
```

## 3- Origin Distribution
### Asian Cars are mostly purchased by the population.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Origin', data=df)
plt.title('Distribution of Car Origins')
plt.show()
```

## 4- DriveTrain Types
### Front Wheel Drive (FWD) is the most common than RWD and 4WD.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='DriveTrain', data=df)
plt.title('Distribution of DriveTrain Types')
plt.show()
```

### converting Dollar symbol columns to float type.

```python
df['MSRP'] = df['MSRP'].replace('[\$,]', '', regex=True).astype(float)
df['Invoice'] = df['Invoice'].replace('[\$,]', '', regex=True).astype(float)
df
```

## 5- MSRP and Invoice Analysis

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='MSRP', y='Invoice', data=df , hue='Type' , palette='viridis', s=80)
plt.title('MSRP and Invoice Prices')
plt.legend(title='Car Type')
plt.show()
```

## 6- Engine Characteristics
### On average 3200cc of engine size is preffered with 6 cylinders giving 200 Hp apx.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(df['EngineSize'], bins=20, kde=True, ax=axes[0])
sns.countplot(x='Cylinders', data=df, ax=axes[1])
sns.histplot(df['Horsepower'], bins=20, kde=True, ax=axes[2])
fig.suptitle('Distribution of Engine Characteristics')
plt.show()
```

```python
filtered_df = df.groupby('Make').filter(lambda x: (x['Cylinders'] == 6).any())
mean_horsepower = filtered_df.groupby('Make')['EngineSize'].mean()
mean_horsepower.mean()
```

## 7- Fuel Efficiency
### Obviously Hybrid cars will give higher mileage over other cars.

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='MPG_City', y='MPG_Highway', hue='Type', data=df, palette='viridis', s=80)
plt.title('City MPG vs. Highway MPG with Car Type Differentiation')
plt.legend(title='Car Type')
plt.show()
```

```python
df[df.Type == 'Hybrid']
```

## 8- Weight vs. Length
### Truck being the heaviest and longest vehicle, but some cars have higher dimensions than trucks.

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Weight', y='Length', data=df , hue='Type' , palette='viridis' , s=80)
plt.title('Weight vs. Length of Cars')
plt.show()
```

```python
df[df['Weight'] > 5900]
```

## 9- Correlation between horsepower and MPG in city

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Horsepower', y='MPG_City', data=df , hue='Type' , palette='viridis' , s=80)
plt.title('Scatter Plot of Horsepower vs. MPG_City')
plt.xlabel('Horsepower')
plt.ylabel('MPG_City')
plt.show()
```

## 10- Top Models by Price and Horsepower:

```python
top_models_by_msrp = df.nlargest(10, 'MSRP')[['Make', 'Model', 'MSRP']]
top_models_by_horsepower = df.nlargest(10, 'Horsepower')[['Make', 'Model', 'Horsepower']]

print("Top Models by MSRP:\n", top_models_by_msrp)
print("\nTop Models by Horsepower:\n", top_models_by_horsepower)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_models_by_msrp['Model'], top_models_by_msrp['MSRP'], color='blue', label='MSRP')
plt.xlabel('Car Model')
plt.ylabel('MSRP')
plt.title('Top 10 Models by MSRP')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()

# Bar plot for top models by Horsepower
plt.figure(figsize=(12, 6))
plt.bar(top_models_by_horsepower['Model'], top_models_by_horsepower['Horsepower'], color='green', label='Horsepower')
plt.xlabel('Car Model')
plt.ylabel('Horsepower')
plt.title('Top 10 Models by Horsepower')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()
```
